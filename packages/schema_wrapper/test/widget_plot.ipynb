```python
import pandas as pd
from mosaic_widget import MosaicWidget
```

```python
weather = pd.read_csv("https://uwdata.github.io/mosaic-datasets/data/seattle-weather.csv", parse_dates=['date'])
weather
```

```python
spec = {
    'colorDomain': '$domain', 'colorRange': '$colors', 'plot': [{'data': {'filterBy': '$click', 'from': 'weather'}, 'fill': 'weather', 'mark': 'dot', 'r': 'precipitation', 'x': {'dateMonthDay': 'date'}, 'y': 'temp_max'}], 'rDomain': 'Fixed', 'rRange': [2, 10], 'width': 800, 'xTickFormat': '%b', 'xyDomain': 'Fixed'
}
```

```python
MosaicWidget(spec, data = {"weather": weather})
```

```python
spec = {'height': 200, 'plot': [{'data': {'from': 'aapl'}, 'mark': 'lineY', 'x': 'Date', 'y': 'Close'}], 'width': 680}
```

```python
stocks = pd.read_csv("https://uwdata.github.io/mosaic-datasets/data/stocks.csv", parse_dates=['Date'])
stocks
```

```python
# Filter for AAPL records
aapl = stocks[stocks['Symbol'] == 'AAPL']
aapl
```

```python
# Create the widget with the filtered data
MosaicWidget(spec, data={"aapl": aapl})
```
